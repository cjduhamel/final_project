```python
import pandas as pd
splits = {'train': 'train.jsonl', 'validation': 'dev.jsonl', 'test': 'test.jsonl'}
df = pd.read_json("hf://datasets/copenlu/citeworth/" + splits["train"], lines=True)

df.head()
```

```python
first_entry = df.iloc[0]
print(first_entry)

print(first_entry['original_text'])


```

```python
import requests
from dotenv import load_dotenv
import os
load_dotenv()

API_KEY = os.getenv("SEMANTIC_SCHOLAR_API_KEY")
# Choose either DOI, arXiv ID, or Semantic Scholar paper ID
paper_id = first_entry['paper_id']  # example: DOI

url = f"https://api.semanticscholar.org/graph/v1/paper/CorpusId:{paper_id}"
params = {
    "fields": "title,abstract,authors,year,venue,url"
}

response = requests.get(url, params=params, headers={"x-api-key": API_KEY})

if response.status_code == 200:
    paper = response.json()
    print("Title:", paper["title"])
    print("Authors:", [author["name"] for author in paper["authors"]])
    print("Year:", paper["year"])
    print("Venue:", paper["venue"])
    print("Abstract:", paper["abstract"])
    print("URL:", paper["url"])
else:
    print("Failed to fetch paper:", response.status_code, response.text)
```
